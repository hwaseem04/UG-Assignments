```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Swarm_Behaviour.csv")
```

```python
data.info()
```

```python
data.head()
```

No null values

```python
print(data.isna().any().sum())
```

```python
for i in data:
    print(np.ptp(data[i]))
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(data.iloc[:, :], train_size=0.8, random_state=2)
```

```python
test.shape
```

```python
train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
y_train = train.iloc[:, -1]
X_train = sc.fit_transform(train.iloc[:, :-1])
```

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression()
model = reg.fit(X_train, y_train)
```

```python
y_test = test.iloc[:, -1]
X_test  = test.iloc[:, :-1]
```

```python
X_test = sc.transform(X_test)
```

```python
pred = model.predict(X_test)
```

```python
(pred==y_test).mean()
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
ngs = KNeighborsClassifier()
model = ngs.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
(pred == y_test).mean()
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb = GaussianNB()
model = nb.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
(pred == y_test).mean()
```

### Result:

--> KNN has better accuracy - 88.78%
